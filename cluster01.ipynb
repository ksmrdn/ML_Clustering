### Clustering
## K-Means Clustering
- Unsupervised Learning
- Unlabeled Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
df = pd.DataFrame(iris['data'], columns=['SL','SW','PL','PW'])
df['target'] = iris['target']
df
```

```python
plt.subplot(121)
plt.plot(df[['SL']],df['SW'], 'r.')
plt.subplot(122)
plt.plot(df[['PL']],df['PW'], 'r.')
```

Elbow Cluster

```python
km = KMeans(n_clusters=3, max_iter = 5000)
km.fit(df[['SL','SW','PL','PW']])
```

```python
km.labels_
```

```python
#model.labels_
df['pred'] = km.predict(df[['SL','SW','PL','PW']])
df
# Label target asli dan label predict tidak sama
#1. Target asli mengalah 0 1 2 == 1 0 2
#2. Tingkatkan iterasi 
```

```python
plt.subplot(221)
plt.plot(df['SL'][df['target']==0],df['SW'][df['target']==0],'k.')
plt.plot(df['SL'][df['target']==1],df['SW'][df['target']==1],'c.')
plt.plot(df['SL'][df['target']==2],df['SW'][df['target']==2],'y.')
plt.subplot(222)
plt.plot(df['PL'][df['target']==0],df['PW'][df['target']==0],'k.')
plt.plot(df['PL'][df['target']==1],df['PW'][df['target']==1],'c.')
plt.plot(df['PL'][df['target']==2],df['PW'][df['target']==2],'y.')

plt.subplot(223)
plt.plot(df['SL'][df['pred']==0],df['SW'][df['pred']==0],'k.')
plt.plot(df['SL'][df['pred']==1],df['SW'][df['pred']==1],'c.')
plt.plot(df['SL'][df['pred']==2],df['SW'][df['pred']==2],'y.')
plt.subplot(224)
plt.plot(df['PL'][df['pred']==0],df['PW'][df['pred']==0],'k.')
plt.plot(df['PL'][df['pred']==1],df['PW'][df['pred']==1],'c.')
plt.plot(df['PL'][df['pred']==2],df['PW'][df['pred']==2],'y.')
```

```python
#centroiods
km.cluster_centers_
```

```python
#check for the best cluster
k = range(1,15)
inertia = []

for i in k:
    km1 = KMeans(n_clusters= i)
    km1.fit(df[['SL','SW','PL','PW']])
    inertia.append(km1.inertia_)
print(k)
print(inertia)
```

```python
plt.plot(k, inertia)
plt.xlabel('K Value')
plt.ylabel('Inertia Sum Squared Error')
plt.plot(k[2], inertia[2],'ro')
plt.annotate('Best k value = 3', xy=(k[2], inertia[2]), xytext=(4,200), arrowprops=dict(facecolor='k', shrink=0.2))
```

### Evaluation Metrics
- tahu atau pura-pura tahu jumlah target

```python
from sklearn import metrics
```

```python
metrics.cluster.contingency_matrix(df['target'], df['pred'])
```

```python
metrics.confusion_matrix(df['target'], df['pred'])
```

```python
#adjusted rand index => similarity
metrics.adjusted_rand_score(df['target'],df['pred'])
```

```python
#Mutual Information Score => Agreement
print(metrics.adjusted_mutual_info_score(df['target'],df['pred']))
```

```python
#Homogenitas => Cluster tersebut berapa persen dimasuki oleh anggota cluster tersebut degan tepat
metrics.homogeneity_score(df['target'], df['pred'])
```

```python
#Komplisitas => Seberapa banyak anggota kelas masuk ke anggota kelas yang benar
metrics.completeness_score(df['target'], df['pred'])
```

```python
#rata2 harmonis homogenitas & komplisitas
metrics.v_measure_score(df['target'], df['pred'])
```

```python
metrics.homogeneity_completeness_v_measure(df['target'], df['pred'])
```

```python
# Fowlkes-Mallows Score => TP/ sqrt((TP+FP)*(TP+FN))
metrics.fowlkes_mallows_score(df['target'], df['pred'])
```

### Unknown Labels

```python
# Silhoutte Coeficient
#-1 s/d 1 => Seberapa bagus model mendefinisikan cluster
metrics.silhouette_score(df[['SL','SW','PL','PW']], km1.labels_)
```

```python
#calinksi-Harabasz Index: Variance Ratio Criterion
#Seberapa bagus model mendefinisikan cluster
#Untuk dibandingkan dengan model ini
chi = metrics.calinski_harabasz_score(df[['SL','SW','PL','PW']], km1.labels_)
```

```python
metrics.davies_bouldin_score(df[['SL','PL','PL','PW']], km1.labels_)
```
